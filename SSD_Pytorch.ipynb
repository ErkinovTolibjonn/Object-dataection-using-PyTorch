```python
ig = cv2.imread('/content/000000037777.jpg')
for i in range(num):
  x1,y1,x2,y2 = boxes[i].numpy().astype('int')
  ig = cv2.rectangle(ig,(x1,y1),(x2,y2),(0,0,255),2)
  class_name = coco_names[labels.numpy()[i]-1]
  ig = cv2.putText(ig,class_name,(x1,y1-10),font,0.5,(255,0,0),1,cv2.LINE_AA)
```
